## 第一部分：面向对象基础概念（30 分钟）

### 1.1 什么是面向对象编程 (OOP)

面向对象编程是一种编程范式，它将现实世界的事物抽象为程序中的对象。

**核心概念：**
- **对象 (Object)**: 现实世界实体的抽象，包含数据和方法
- **类 (Class)**: 对象的模板或蓝图，定义对象的结构
- **实例 (Instance)**: 类的具体实现，一个真实存在的对象

**在 RoboMaster 中的对象思维：**
- 机器人 → 对象
- 机器人类型（Hero、Infantry等） → 类
- 具体的 Hero-1 机器人 → 实例

```python
# 第一个完整的类定义 - 基于真实RoboMaster规则

print("=== 面向对象编程基础 ===")

class Robot:
    """机器人基类 - 基于真实RoboMaster规则设计"""
    
    # 类属性（所有实例共享）
    game_version = "2024"
    robot_count = 0
    
    def __init__(self, name, robot_type, hp):
        """
        构造函数：创建机器人实例时自动调用
        
        Args:
            name (str): 机器人名称
            robot_type (str): 机器人类型 (Hero/Infantry/Engineer)
            hp (int): 生命值
        """
        # 实例属性（每个实例独有）
        self.name = name
        self.robot_type = robot_type
        self.hp = hp
        self.max_hp = hp
        self.position = (0, 0)
        self.status = "待命"
        
        # 根据机器人类型设置攻击力
        if robot_type == "Hero":
            self.damage = 10  # 英雄机器人大弹丸，10点伤害
        elif robot_type == "Infantry":
            self.damage = 1   # 步兵机器人小弹丸，1点伤害
        else:  # Engineer
            self.damage = 0   # 工程机器人无攻击能力
        
        # 更新类属性
        Robot.robot_count += 1
        
        print(f"✅ 创建{robot_type}机器人: {name} (血量:{hp}HP, 攻击:{self.damage}点)")
    
    def __str__(self):
        """字符串表示 - 用于print()"""
        return f"{self.robot_type}({self.name}, HP:{self.hp}/{self.max_hp}, 攻击:{self.damage}点)"
    
    def __repr__(self):
        """开发者表示 - 用于调试"""
        return f"Robot('{self.name}', '{self.robot_type}', {self.hp})"
    
    # 实例方法
    def move(self, x, y):
        """移动机器人到指定位置"""
        if self.status == "待命":
            self.position = (x, y)
            print(f"{self.name} 移动到位置 ({x}, {y})")
        else:
            print(f"❌ {self.name} 状态异常，无法移动")
    
    def attack(self, target):
        """攻击目标机器人"""
        if self.damage == 0:
            print(f"❌ {self.name} 是工程机器人，无法攻击")
            return 0
        
        if self.status != "待命":
            print(f"❌ {self.name} 状态异常，无法攻击")
            return 0
        
        print(f"🔥 {self.name} 攻击 {target.name}，造成 {self.damage} 点伤害")
        target.take_damage(self.damage)
        return self.damage
    
    def take_damage(self, damage):
        """受到伤害"""
        old_hp = self.hp
        self.hp = max(0, self.hp - damage)
        print(f"💥 {self.name} 受到 {damage} 点伤害: {old_hp} -> {self.hp}")
        
        if self.hp == 0:
            self.status = "损坏"
            print(f"� {self.name} 已被摧毁!")
    
    def repair(self, heal_amount):
        """修复机器人"""
        if self.status == "损坏":
            print(f"❌ {self.name} 已损坏，需要完全重建")
            return
        
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + heal_amount)
        print(f"🔧 {self.name} 修复 {heal_amount} 点: {old_hp} -> {self.hp}")
    
    def mine_ore(self):
        """兑矿操作（仅工程机器人可用）"""
        if self.robot_type != "Engineer":
            print(f"❌ {self.name} 不是工程机器人，无法兑矿")
            return False
        
        if self.status != "待命":
            print(f"❌ {self.name} 状态异常，无法兑矿")
            return False
        
        print(f"⛏️ {self.name} 执行兑矿操作")
        return True
    
    def get_info(self):
        """获取机器人详细信息"""
        return {
            "name": self.name,
            "type": self.robot_type,
            "hp": f"{self.hp}/{self.max_hp}",
            "damage": self.damage,
            "position": self.position,
            "status": self.status
        }

# 创建不同类型的机器人
print("\n=== 创建机器人队伍 ===")
hero_robot = Robot("Hero-1", "Hero", 600)      # 英雄机器人：血量厚，攻击强
infantry_1 = Robot("Infantry-1", "Infantry", 200)  # 步兵机器人：标准血量，攻击弱
infantry_2 = Robot("Infantry-2", "Infantry", 200)  # 第二个步兵
engineer = Robot("Engineer-1", "Engineer", 250)    # 工程机器人：中等血量，无攻击

# 显示所有机器人信息
print(f"\n=== 机器人信息总览 ===")
robots = [hero_robot, infantry_1, infantry_2, engineer]
for robot in robots:
    print(f"  {robot}")

print(f"\n总共创建了 {Robot.robot_count} 个机器人")

# 演示基本操作
print(f"\n=== 基本操作演示 ===")

# 移动测试
hero_robot.move(5, 3)
engineer.move(2, 1)

# 攻击测试
print(f"\n--- 攻击测试 ---")
hero_robot.attack(infantry_1)     # 英雄攻击步兵
infantry_2.attack(hero_robot)     # 步兵攻击英雄
engineer.attack(infantry_1)       # 工程试图攻击（应该失败）

# 兑矿测试
print(f"\n--- 特殊能力测试 ---")
engineer.mine_ore()               # 工程兑矿
hero_robot.mine_ore()             # 英雄试图兑矿（应该失败）

# 修复测试
print(f"\n--- 修复测试 ---")
hero_robot.repair(50)             # 修复英雄
infantry_1.repair(100)            # 修复步兵

# 查看最终状态
print(f"\n=== 最终状态 ===")
for robot in robots:
    info = robot.get_info()
    print(f"  {info['name']}: {info['type']}, {info['hp']}HP, {info['damage']}攻击, 状态:{info['status']}")

print("\n✅ 基础类定义和对象操作完成！")
```

## 第二部分：面向对象三大特性（45 分钟）

### 2.1 封装 (Encapsulation)

封装是将数据和操作数据的方法绑定在一起，并隐藏内部实现细节的特性。

```python
# 封装和继承示例 - 基于真实RoboMaster规则

print("=== 面向对象三大特性演示 ===")

# 基类：机器人
class Robot:
    """基础机器人类 - 统一的机器人接口"""
    
    def __init__(self, name, robot_type, hp):
        self.name = name
        self.robot_type = robot_type
        self._hp = hp  # 受保护属性（使用下划线表示）
        self._max_hp = hp
        self.position = (0, 0)
        self.status = "待命"
    
    @property
    def hp(self):
        """血量属性（只读）"""
        return self._hp
    
    @hp.setter 
    def hp(self, value):
        """设置血量，自动限制在合理范围内"""
        self._hp = max(0, min(value, self._max_hp))
        if self._hp == 0:
            self.status = "损坏"
    
    @property
    def max_hp(self):
        """最大血量属性（只读）"""
        return self._max_hp
    
    def move(self, x, y):
        """移动方法"""
        if self.status == "待命":
            self.position = (x, y)
            print(f"{self.name} 移动到 ({x}, {y})")
        else:
            print(f"❌ {self.name} 无法移动，状态: {self.status}")
    
    def attack(self, target):
        """基础攻击方法（子类应重写）"""
        print(f"{self.name} 对 {target} 执行基础攻击")
        return 0
    
    def take_damage(self, damage):
        """受到伤害"""
        old_hp = self.hp
        self.hp -= damage
        print(f"💥 {self.name} 受到 {damage} 伤害: {old_hp} -> {self.hp}")

# 派生类：英雄机器人
class HeroRobot(Robot):
    """英雄机器人类 - 血量厚，大弹丸攻击"""
    
    def __init__(self, name):
        super().__init__(name, "Hero", 600)  # 英雄血量600
        self.weapon = "42mm大弹丸"
        self.armor_level = "重装甲"
        self.damage_per_shot = 10
    
    def attack(self, target):
        """重写攻击方法 - 大弹丸攻击"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法攻击，状态: {self.status}")
            return 0
        
        print(f"🔥 {self.name} 使用 {self.weapon} 攻击 {target.name}!")
        target.take_damage(self.damage_per_shot)
        return self.damage_per_shot
    
    def charge_attack(self, target):
        """英雄特有技能 - 蓄力攻击"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法使用特殊攻击")
            return 0
        
        charge_damage = self.damage_per_shot * 2  # 蓄力伤害翻倍
        print(f"⚡ {self.name} 发动蓄力攻击!")
        target.take_damage(charge_damage)
        return charge_damage

# 派生类：步兵机器人  
class InfantryRobot(Robot):
    """步兵机器人类 - 标准血量，小弹丸攻击"""
    
    def __init__(self, name):
        super().__init__(name, "Infantry", 200)  # 步兵血量200
        self.weapon = "17mm小弹丸"
        self.mobility = "高机动性"
        self.damage_per_shot = 1
    
    def attack(self, target):
        """重写攻击方法 - 小弹丸攻击"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法攻击，状态: {self.status}")
            return 0
        
        print(f"🔫 {self.name} 使用 {self.weapon} 攻击 {target.name}!")
        target.take_damage(self.damage_per_shot)
        return self.damage_per_shot
    
    def rapid_fire(self, target, shots=3):
        """步兵特有技能 - 连发射击"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法使用连发射击")
            return 0
        
        total_damage = 0
        print(f"🔫🔫🔫 {self.name} 对 {target.name} 进行 {shots} 连发射击!")
        for i in range(shots):
            target.take_damage(self.damage_per_shot)
            total_damage += self.damage_per_shot
        return total_damage

# 派生类：工程机器人
class EngineerRobot(Robot):
    """工程机器人类 - 中等血量，无攻击能力，可兑矿"""
    
    def __init__(self, name):
        super().__init__(name, "Engineer", 250)  # 工程血量250
        self.tool = "兑矿装置"
        self.ore_capacity = 0
        self.max_ore_capacity = 100
    
    def attack(self, target):
        """工程机器人无攻击能力"""
        print(f"❌ {self.name} 是工程机器人，无法攻击")
        return 0
    
    def mine_ore(self, amount=20):
        """兑矿操作"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法兑矿，状态: {self.status}")
            return 0
        
        if self.ore_capacity >= self.max_ore_capacity:
            print(f"❌ {self.name} 矿石容量已满")
            return 0
        
        actual_amount = min(amount, self.max_ore_capacity - self.ore_capacity)
        self.ore_capacity += actual_amount
        print(f"⛏️ {self.name} 兑取 {actual_amount} 单位矿石 (总量: {self.ore_capacity}/{self.max_ore_capacity})")
        return actual_amount
    
    def repair_ally(self, target, heal_amount=50):
        """修复盟友"""
        if self.status != "待命":
            print(f"❌ {self.name} 无法修复，状态: {self.status}")
            return 0
        
        if target.status == "损坏":
            print(f"❌ {target.name} 已损坏，无法修复")
            return 0
        
        old_hp = target.hp
        target.hp = min(target.max_hp, target.hp + heal_amount)
        actual_heal = target.hp - old_hp
        print(f"🔧 {self.name} 修复 {target.name}: {old_hp} -> {target.hp} (+{actual_heal})")
        return actual_heal

# 创建机器人队伍
print("\n=== 创建专业机器人队伍 ===")
hero = HeroRobot("Hero-Alpha")
infantry1 = InfantryRobot("Infantry-Beta")  
infantry2 = InfantryRobot("Infantry-Gamma")
engineer = EngineerRobot("Engineer-Delta")

robots = [hero, infantry1, infantry2, engineer]

# 显示初始状态
print(f"\n=== 队伍初始状态 ===")
for robot in robots:
    print(f"  {robot.name}: {robot.robot_type}, {robot.hp}/{robot.max_hp}HP")

# 演示不同攻击方式
print(f"\n=== 战斗演示 ===")

# 英雄攻击步兵
hero.attack(infantry1)
hero.charge_attack(infantry1)  # 特殊攻击

# 步兵反击
infantry1.attack(hero)
infantry1.rapid_fire(hero, 5)  # 连发攻击

# 工程机器人尝试攻击（应该失败）
engineer.attack(hero)

# 演示特殊能力
print(f"\n=== 特殊能力演示 ===")
engineer.mine_ore(30)
engineer.mine_ore(50)
engineer.repair_ally(infantry1, 80)

# 显示最终状态
print(f"\n=== 战斗后状态 ===")
for robot in robots:
    status_info = f"{robot.name}: {robot.robot_type}, {robot.hp}/{robot.max_hp}HP"
    if hasattr(robot, 'ore_capacity'):
        status_info += f", 矿石: {robot.ore_capacity}"
    print(f"  {status_info}")

print(f"\n✅ 继承和多态演示完成！")
```

## 第三部分：异常处理（20 分钟）

### 3.1 异常的基本概念

异常是程序运行时发生的错误，Python提供了完善的异常处理机制来处理这些错误。

```python
# 异常处理详解

print("=== 异常处理机制 ===")

# 自定义异常类
class RobotError(Exception):
    """机器人相关异常的基类"""
    pass

class RobotDestroyedException(RobotError):
    """机器人已损坏异常"""
    pass

class InsufficientEnergyException(RobotError):
    """能量不足异常"""
    pass

class InvalidPositionException(RobotError):
    """无效位置异常"""
    pass

class RobotControlSystem:
    """机器人控制系统 - 异常处理示例"""
    
    def __init__(self):
        self.robots = {}
        self.field_size = (1000, 1000)
    
    def add_robot(self, robot_id, name, robot_type):
        """添加机器人"""
        try:
            if robot_id in self.robots:
                raise ValueError(f"机器人ID {robot_id} 已存在")
            
            hp_map = {"Hero": 400, "Infantry": 200, "Engineer": 250}
            if robot_type not in hp_map:
                raise ValueError(f"不支持的机器人类型: {robot_type}")
            
            self.robots[robot_id] = {
                "name": name,
                "type": robot_type,
                "hp": hp_map[robot_type],
                "max_hp": hp_map[robot_type],
                "energy": 100,
                "position": (0, 0),
                "status": "正常"
            }
            
            print(f"✅ 成功添加机器人: {name} (ID: {robot_id})")
            
        except ValueError as e:
            print(f"❌ 添加机器人失败: {e}")
            raise
    
    def move_robot(self, robot_id, x, y):
        """移动机器人"""
        try:
            # 检查机器人是否存在
            if robot_id not in self.robots:
                raise KeyError(f"机器人ID {robot_id} 不存在")
            
            robot = self.robots[robot_id]
            
            # 检查机器人状态
            if robot["status"] == "损坏":
                raise RobotDestroyedException(f"机器人 {robot['name']} 已损坏，无法移动")
            
            # 检查位置有效性
            if not (0 <= x <= self.field_size[0] and 0 <= y <= self.field_size[1]):
                raise InvalidPositionException(f"位置 ({x}, {y}) 超出场地范围")
            
            # 检查能量
            energy_cost = 10
            if robot["energy"] < energy_cost:
                raise InsufficientEnergyException(f"能量不足，需要 {energy_cost}，当前 {robot['energy']}")
            
            # 执行移动
            old_pos = robot["position"]
            robot["position"] = (x, y)
            robot["energy"] -= energy_cost
            
            print(f"✅ {robot['name']} 从 {old_pos} 移动到 ({x}, {y})，消耗能量 {energy_cost}")
            
        except (KeyError, RobotDestroyedException, InvalidPositionException, InsufficientEnergyException) as e:
            print(f"❌ 移动失败: {e}")
            return False
        except Exception as e:
            print(f"❌ 意外错误: {e}")
            return False
        
        return True
    
    def attack_robot(self, attacker_id, target_id):
        """机器人攻击"""
        try:
            # 检查攻击者和目标
            if attacker_id not in self.robots:
                raise KeyError(f"攻击者 {attacker_id} 不存在")
            if target_id not in self.robots:
                raise KeyError(f"目标 {target_id} 不存在")
            
            attacker = self.robots[attacker_id]
            target = self.robots[target_id]
            
            # 检查攻击者状态
            if attacker["status"] == "损坏":
                raise RobotDestroyedException(f"攻击者 {attacker['name']} 已损坏")
            
            # 检查目标状态
            if target["status"] == "损坏":
                raise RobotDestroyedException(f"目标 {target['name']} 已损坏")
            
            # 检查能量
            if attacker["energy"] < 20:
                raise InsufficientEnergyException("攻击需要20点能量")
            
            # 执行攻击
            damage = 50
            attacker["energy"] -= 20
            target["hp"] -= damage
            
            print(f"💥 {attacker['name']} 攻击 {target['name']}，造成 {damage} 伤害")
            
            # 检查目标是否被摧毁
            if target["hp"] <= 0:
                target["hp"] = 0
                target["status"] = "损坏"
                print(f"💀 {target['name']} 被摧毁!")
            
        except Exception as e:
            print(f"❌ 攻击失败: {e}")
            return False
        
        return True
    
    def get_robot_status(self, robot_id):
        """获取机器人状态"""
        try:
            if robot_id not in self.robots:
                raise KeyError(f"机器人ID {robot_id} 不存在")
            
            return self.robots[robot_id].copy()
            
        except KeyError as e:
            print(f"❌ 查询失败: {e}")
            return None

# 异常处理测试
print("\\n=== 异常处理测试 ===")

system = RobotControlSystem()

# 测试正常操作
try:
    system.add_robot("R001", "Hero-1", "Hero")
    system.add_robot("R002", "Infantry-1", "Infantry")
    print("\\n正常操作完成")
except Exception as e:
    print(f"操作失败: {e}")

# 测试异常情况
print(f"\\n=== 异常情况测试 ===")

# 1. 重复ID异常
try:
    system.add_robot("R001", "Hero-2", "Hero")
except ValueError:
    print("✅ 成功捕获重复ID异常")

# 2. 无效机器人类型
try:
    system.add_robot("R003", "Unknown", "Tank")
except ValueError:
    print("✅ 成功捕获无效类型异常")

# 3. 移动到无效位置
try:
    system.move_robot("R001", 2000, 3000)
except:
    print("✅ 成功捕获无效位置异常")

# 4. 能量不足（通过多次移动消耗能量）
print(f"\\n=== 能量消耗测试 ===")
for i in range(12):  # 尝试移动12次，每次消耗10能量
    success = system.move_robot("R001", i*10, i*10)
    if not success:
        print(f"第{i+1}次移动失败")
        break

# 5. 攻击已损坏的机器人
print(f"\\n=== 战斗测试 ===")
# 先攻击直到目标被摧毁
for i in range(10):  # 多次攻击
    success = system.attack_robot("R001", "R002")
    if not success:
        break

# 尝试攻击已损坏的机器人
system.attack_robot("R001", "R002")

print("\\n✅ 异常处理学习完成！")
```

## 第四部分：综合实践项目 - 机器人控制系统模拟器

让我们创建一个完整的机器人控制系统，综合运用所有学到的知识。

```python
# 综合实践项目：机器人控制系统模拟器 - 基于真实RoboMaster规则

import random
from datetime import datetime
from enum import Enum

print("🤖 RoboMaster 机器人控制系统模拟器")
print("=" * 60)

class RobotState(Enum):
    """机器人状态枚举"""
    STANDBY = "待命"
    ACTIVE = "激活"
    MOVING = "移动中" 
    FIGHTING = "战斗中"
    MINING = "兑矿中"
    DESTROYED = "已损坏"

class BattleRobot:
    """战斗机器人基类 - 实现真实RoboMaster规则"""
    
    def __init__(self, robot_id, name, robot_type):
        self.robot_id = robot_id
        self.name = name
        self.robot_type = robot_type
        self.state = RobotState.STANDBY
        self.position = (0, 0)
        self.energy = 100
        self.max_energy = 100
        
        # 根据机器人类型设置属性（真实规则）
        if robot_type == "Hero":
            self.hp = 600
            self.max_hp = 600
            self.damage = 10  # 大弹丸
            self.energy_per_move = 8
            self.energy_per_attack = 15
            self.can_attack = True
            self.can_mine = False
        elif robot_type == "Infantry":
            self.hp = 200
            self.max_hp = 200
            self.damage = 1   # 小弹丸
            self.energy_per_move = 5
            self.energy_per_attack = 8
            self.can_attack = True
            self.can_mine = False
        elif robot_type == "Engineer":
            self.hp = 250
            self.max_hp = 250
            self.damage = 0   # 无攻击能力
            self.energy_per_move = 6
            self.energy_per_attack = 0
            self.can_attack = False
            self.can_mine = True
            self.ore_count = 0
            self.max_ore = 100
        else:
            raise ValueError(f"未知的机器人类型: {robot_type}")
        
        self.battle_log = []
        
    def log_action(self, action):
        """记录操作日志"""
        timestamp = datetime.now().strftime("%H:%M:%S")
        log_entry = f"[{timestamp}] {self.name}: {action}"
        self.battle_log.append(log_entry)
        print(log_entry)
    
    def move(self, x, y):
        """移动机器人"""
        if self.state == RobotState.DESTROYED:
            self.log_action("❌ 已损坏，无法移动")
            return False
        
        if self.energy < self.energy_per_move:
            self.log_action(f"❌ 能量不足，需要 {self.energy_per_move} 点能量")
            return False
        
        old_pos = self.position
        self.position = (x, y)
        self.energy -= self.energy_per_move
        self.state = RobotState.MOVING
        
        self.log_action(f"移动: {old_pos} -> {self.position}, 消耗 {self.energy_per_move} 能量")
        return True
    
    def attack(self, target):
        """攻击目标"""
        if not self.can_attack:
            self.log_action(f"❌ {self.robot_type} 无攻击能力")
            return False
        
        if self.state == RobotState.DESTROYED:
            self.log_action("❌ 已损坏，无法攻击")
            return False
        
        if self.energy < self.energy_per_attack:
            self.log_action(f"❌ 能量不足，需要 {self.energy_per_attack} 点能量")
            return False
        
        if target.state == RobotState.DESTROYED:
            self.log_action(f"❌ 目标 {target.name} 已被摧毁")
            return False
        
        # 执行攻击
        self.energy -= self.energy_per_attack
        self.state = RobotState.FIGHTING
        
        # 计算伤害（可能有暴击）
        damage = self.damage
        is_critical = random.random() < 0.1  # 10%暴击率
        if is_critical:
            damage *= 2
            self.log_action(f"🔥 暴击! 对 {target.name} 造成 {damage} 点伤害")
        else:
            self.log_action(f"🔫 对 {target.name} 造成 {damage} 点伤害")
        
        # 目标受伤
        target.take_damage(damage)
        return True
    
    def take_damage(self, damage):
        """受到伤害"""
        old_hp = self.hp
        self.hp = max(0, self.hp - damage)
        
        if self.hp == 0:
            self.state = RobotState.DESTROYED
            self.log_action(f"💀 受到 {damage} 伤害 ({old_hp} -> {self.hp})，机器人被摧毁!")
        else:
            self.log_action(f"💥 受到 {damage} 伤害 ({old_hp} -> {self.hp})")
    
    def mine_ore(self):
        """兑矿操作（仅工程机器人）"""
        if not self.can_mine:
            self.log_action(f"❌ {self.robot_type} 无法兑矿")
            return False
        
        if self.state == RobotState.DESTROYED:
            self.log_action("❌ 已损坏，无法兑矿")
            return False
        
        if self.ore_count >= self.max_ore:
            self.log_action("❌ 矿石容量已满")
            return False
        
        # 兑矿操作
        ore_gained = random.randint(10, 30)
        self.ore_count = min(self.max_ore, self.ore_count + ore_gained)
        self.state = RobotState.MINING
        
        self.log_action(f"⛏️ 兑取 {ore_gained} 矿石，总计: {self.ore_count}/{self.max_ore}")
        return True
    
    def repair(self, heal_amount):
        """修复（需要消耗矿石）"""
        if self.state == RobotState.DESTROYED:
            self.log_action("❌ 机器人已完全损坏，无法修复")
            return False
        
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + heal_amount)
        actual_heal = self.hp - old_hp
        
        self.log_action(f"🔧 修复 {actual_heal} 点血量 ({old_hp} -> {self.hp})")
        return True
    
    def recharge(self, energy_amount=50):
        """补充能量"""
        old_energy = self.energy
        self.energy = min(self.max_energy, self.energy + energy_amount)
        actual_recharge = self.energy - old_energy
        
        self.log_action(f"🔋 补充 {actual_recharge} 点能量 ({old_energy} -> {self.energy})")
        return actual_recharge
    
    def get_status(self):
        """获取状态信息"""
        status = {
            "name": self.name,
            "type": self.robot_type,
            "hp": f"{self.hp}/{self.max_hp}",
            "energy": f"{self.energy}/{self.max_energy}",
            "damage": self.damage,
            "position": self.position,
            "state": self.state.value
        }
        
        if hasattr(self, 'ore_count'):
            status["ore"] = f"{self.ore_count}/{self.max_ore}"
        
        return status

class BattleSimulator:
    """战斗模拟器"""
    
    def __init__(self):
        self.robots = {}
        self.battle_round = 0
        self.battle_log = []
    
    def add_robot(self, robot_id, name, robot_type):
        """添加机器人"""
        if robot_id in self.robots:
            print(f"❌ 机器人ID {robot_id} 已存在")
            return False
        
        robot = BattleRobot(robot_id, name, robot_type)
        self.robots[robot_id] = robot
        print(f"✅ 添加 {robot_type} 机器人: {name} (ID: {robot_id})")
        return True
    
    def get_robot(self, robot_id):
        """获取机器人"""
        return self.robots.get(robot_id)
    
    def display_all_robots(self):
        """显示所有机器人状态"""
        print(f"\n📊 机器人状态总览 (第 {self.battle_round} 轮)")
        print("-" * 80)
        
        for robot_id, robot in self.robots.items():
            status = robot.get_status()
            ore_info = f", 矿石: {status['ore']}" if 'ore' in status else ""
            print(f"  {robot_id}: {status['name']} ({status['type']}) - "
                  f"HP: {status['hp']}, 能量: {status['energy']}, "
                  f"伤害: {status['damage']}, 状态: {status['state']}{ore_info}")
    
    def simulate_battle(self):
        """模拟一轮战斗"""
        self.battle_round += 1
        print(f"\n⚔️ === 第 {self.battle_round} 轮战斗 ===")
        
        active_robots = [r for r in self.robots.values() if r.state != RobotState.DESTROYED]
        
        if len(active_robots) < 2:
            print("战斗结束，剩余机器人不足2个")
            return False
        
        # 随机选择攻击者和目标
        attacker = random.choice([r for r in active_robots if r.can_attack and r.energy >= r.energy_per_attack])
        if not attacker:
            print("没有可用的攻击机器人")
            return False
        
        targets = [r for r in active_robots if r != attacker and r.state != RobotState.DESTROYED]
        if not targets:
            print("没有可攻击的目标")
            return False
        
        target = random.choice(targets)
        
        # 执行攻击
        attacker.attack(target)
        
        # 工程机器人可能进行兑矿
        engineers = [r for r in active_robots if r.robot_type == "Engineer" and r.state != RobotState.DESTROYED]
        for engineer in engineers:
            if random.random() < 0.3:  # 30%概率兑矿
                engineer.mine_ore()
        
        return True
    
    def run_simulation(self, rounds=5):
        """运行完整模拟"""
        print(f"\n🎮 开始 {rounds} 轮战斗模拟")
        
        for round_num in range(rounds):
            if not self.simulate_battle():
                break
            
            # 每轮后补充少量能量
            for robot in self.robots.values():
                if robot.state != RobotState.DESTROYED:
                    robot.recharge(20)
            
            self.display_all_robots()
            
        print(f"\n🏁 战斗模拟结束")
        
        # 统计最终结果
        survivors = [r for r in self.robots.values() if r.state != RobotState.DESTROYED]
        destroyed = [r for r in self.robots.values() if r.state == RobotState.DESTROYED]
        
        print(f"\n📈 战斗统计:")
        print(f"  存活机器人: {len(survivors)}")
        print(f"  被摧毁机器人: {len(destroyed)}")
        
        if survivors:
            print(f"  存活者:")
            for robot in survivors:
                print(f"    {robot.name} ({robot.robot_type}): {robot.hp}/{robot.max_hp} HP")

# 创建战斗模拟器并运行演示
print("\n=== 创建机器人队伍 ===")
simulator = BattleSimulator()

# 添加不同类型的机器人
simulator.add_robot("H001", "Hero-Alpha", "Hero")
simulator.add_robot("I001", "Infantry-Beta", "Infantry")
simulator.add_robot("I002", "Infantry-Gamma", "Infantry")
simulator.add_robot("E001", "Engineer-Delta", "Engineer")

# 显示初始状态
simulator.display_all_robots()

# 手动演示几个操作
print(f"\n=== 手动操作演示 ===")
hero = simulator.get_robot("H001")
infantry1 = simulator.get_robot("I001")
engineer = simulator.get_robot("E001")

# 移动测试
hero.move(10, 5)
infantry1.move(15, 8)

# 攻击测试
hero.attack(infantry1)
infantry1.attack(hero)

# 工程机器人兑矿
engineer.mine_ore()
engineer.mine_ore()

# 修复测试
engineer_heal_amount = 30
infantry1.repair(engineer_heal_amount)

# 运行自动战斗模拟
simulator.run_simulation(8)

print(f"\n=== 项目知识点总结 ===")
print("✅ 类的设计与继承")
print("✅ 枚举类型的使用")
print("✅ 异常处理机制")
print("✅ 随机数和概率计算")
print("✅ 日期时间处理")
print("✅ 复杂数据结构管理")
print("✅ 面向对象设计原则")
print("✅ 真实业务逻辑建模")

print(f"\n🎉 机器人控制系统模拟器项目完成!")
print("这个项目展示了如何用面向对象编程构建复杂的真实系统！")
```
