### 1.1 列表(List) - 有序可变集合

列表是Python中最常用的数据结构之一，可以存储多个有序的元素，并且可以修改。

```python
# 列表的创建和基本操作

print("=== 列表基础操作 ===")

# 创建列表的不同方式
robots = ["Hero", "Infantry", "Engineer"]  # 三种主要机器人类型
robot_hp = [600, 200, 250]                # 对应的血量
robot_damage = [10, 1, 0]                 # 对应的攻击伤害
mixed_list = ["Hero", 600, True, 10.5]    # 混合数据类型
empty_list = []
range_list = list(range(1, 6))             # [1, 2, 3, 4, 5]

print(f"机器人类型: {robots}")
print(f"机器人血量: {robot_hp}")
print(f"机器人伤害: {robot_damage}")
print(f"混合列表: {mixed_list}")
print(f"空列表: {empty_list}")
print(f"范围列表: {range_list}")

# 列表访问和索引
print(f"\n=== 列表访问 ===")
print(f"第一种机器人: {robots[0]} (血量: {robot_hp[0]}HP, 伤害: {robot_damage[0]}点)")
print(f"最后一种机器人: {robots[-1]} (血量: {robot_hp[-1]}HP, 伤害: {robot_damage[-1]}点)")
print(f"步兵机器人: {robots[1]} (血量: {robot_hp[1]}HP, 伤害: {robot_damage[1]}点)")

# 列表切片
print(f"\n=== 列表切片 ===")
print(f"前两种机器人: {robots[:2]}")
print(f"后两种机器人: {robots[-2:]}")
print(f"血量切片: {robot_hp[1:3]}")

# 列表修改
print(f"\n=== 列表修改 ===")
original_robots = robots.copy()
print(f"原始列表: {original_robots}")

robots[1] = "Infantry-Advanced"  # 修改单个元素
print(f"修改后: {robots}")

# 恢复原始数据
robots = ["Hero", "Infantry", "Engineer"]

# 列表方法
print(f"\n=== 列表方法 ===")
battle_robots = ["Hero"]
print(f"初始战斗队列: {battle_robots}")

# 添加元素
battle_robots.append("Infantry")           # 在末尾添加
battle_robots.insert(1, "Engineer")        # 在指定位置插入
battle_robots.extend(["Infantry", "Infantry"])  # 添加多个元素

print(f"添加后: {battle_robots}")

# 删除元素
removed_robot = battle_robots.pop()        # 删除并返回最后一个
battle_robots.remove("Engineer")           # 删除指定元素
del battle_robots[1]                       # 删除指定位置

print(f"删除后: {battle_robots}")
print(f"被移除的机器人: {removed_robot}")

# 列表信息查询
print(f"\n=== 列表信息 ===")
all_robot_hp = [600, 200, 250, 200, 600, 250]  # 多个机器人的血量
print(f"所有机器人血量: {all_robot_hp}")
print(f"机器人数量: {len(all_robot_hp)}")
print(f"最高血量: {max(all_robot_hp)}")
print(f"最低血量: {min(all_robot_hp)}")
print(f"总血量: {sum(all_robot_hp)}")
print(f"平均血量: {sum(all_robot_hp)/len(all_robot_hp):.1f}")
print(f"英雄机器人(600HP)数量: {all_robot_hp.count(600)}")
print(f"工程机器人血量(250HP)的位置: {all_robot_hp.index(250)}")

# 列表排序
print(f"\n=== 列表排序 ===")
unsorted_hp = [250, 600, 200, 200, 600]
print(f"原始血量: {unsorted_hp}")

sorted_ascending = sorted(unsorted_hp)       # 不改变原列表
sorted_descending = sorted(unsorted_hp, reverse=True)

print(f"血量升序: {sorted_ascending}")
print(f"血量降序: {sorted_descending}")
print(f"原列表不变: {unsorted_hp}")

# 原地排序
unsorted_hp.sort(reverse=True)
print(f"原地降序排列: {unsorted_hp}")

# 列表推导式（高级特性预览）
print(f"\n=== 列表推导式预览 ===")
hp_values = [600, 200, 250]
hp_after_damage = [hp - 50 for hp in hp_values]  # 每个机器人受到50点伤害
high_hp_robots = [hp for hp in hp_values if hp > 250]  # 筛选高血量机器人

print(f"原始血量: {hp_values}")
print(f"受伤后血量: {hp_after_damage}")
print(f"高血量机器人: {high_hp_robots}")

# RoboMaster应用示例
print(f"\n=== RoboMaster 应用示例 ===")
robot_queue = []
robot_specs = {}

# 模拟机器人加入队列
robot_types = ["Hero", "Infantry", "Infantry", "Engineer", "Hero"]
for i, robot_type in enumerate(robot_types, 1):
    robot_name = f"{robot_type}-{i}"
    robot_queue.append(robot_name)
    
    # 根据类型设置规格
    if robot_type == "Hero":
        specs = {"hp": 600, "damage": 10, "ability": "大弹丸攻击"}
    elif robot_type == "Infantry":
        specs = {"hp": 200, "damage": 1, "ability": "小弹丸攻击"}
    else:  # Engineer
        specs = {"hp": 250, "damage": 0, "ability": "兑矿操作"}
    
    robot_specs[robot_name] = specs

print(f"机器人队列: {robot_queue}")

# 显示机器人详细信息
for robot_name in robot_queue:
    specs = robot_specs[robot_name]
    print(f"  {robot_name}: {specs['hp']}HP, {specs['ability']}, {specs['damage']}点伤害")

# 模拟战斗分配
combat_robots = [name for name in robot_queue if robot_specs[name]['damage'] > 0]
support_robots = [name for name in robot_queue if robot_specs[name]['damage'] == 0]

print(f"\n战斗机器人: {combat_robots}")
print(f"支援机器人: {support_robots}")

print("\n✅ 列表操作学习完成！")
```

### 1.2 字典(Dict) - 键值对映射

字典是Python中非常重要的数据结构，用于存储键值对，类似于现实中的字典。

```python
# 字典的创建和基本操作

print("=== 字典基础操作 ===")

# 创建字典的不同方式 - 基于真实RoboMaster规则
robot_config = {
    "name": "Hero",
    "type": "英雄机器人",
    "hp": 600,           # 英雄机器人血量厚
    "damage": 10,        # 大弹丸攻击力
    "weapon": "42mm大弹丸",
    "active": True
}

# 另一种创建方式
team_scores = dict(red=150, blue=120, green=80)

# 空字典
empty_dict = {}
empty_dict2 = dict()

print(f"英雄机器人配置: {robot_config}")
print(f"队伍得分: {team_scores}")
print(f"空字典: {empty_dict}")

# 字典访问
print(f"\n=== 字典访问 ===")
print(f"机器人名称: {robot_config['name']}")
print(f"机器人血量: {robot_config['hp']}")
print(f"攻击伤害: {robot_config['damage']}")
print(f"武器类型: {robot_config['weapon']}")

# 安全访问（使用get方法）
armor = robot_config.get("armor", "无护甲")  # 如果键不存在，返回默认值
shield = robot_config.get("shield")          # 如果键不存在，返回None

print(f"护甲信息: {armor}")
print(f"护盾信息: {shield}")

# 字典修改
print(f"\n=== 字典修改 ===")
print(f"修改前HP: {robot_config['hp']}")

robot_config["hp"] = 550                     # 受伤后血量
robot_config["armor"] = "Level-3"            # 添加新键
robot_config["status"] = "战斗中"             # 添加状态

print(f"修改后配置: {robot_config}")

# 字典方法
print(f"\n=== 字典方法 ===")
battle_data = {
    "kills": 15,
    "deaths": 3,
    "assists": 8,
    "damage_dealt": 5420,
    "damage_taken": 2100
}

print(f"战斗数据: {battle_data}")
print(f"所有键: {list(battle_data.keys())}")
print(f"所有值: {list(battle_data.values())}")
print(f"键值对: {list(battle_data.items())}")

# 更新字典
new_stats = {"score": 2500, "mvp": True, "accuracy": 0.85}
battle_data.update(new_stats)
print(f"更新后: {battle_data}")

# 删除字典元素
removed_deaths = battle_data.pop("deaths", 0)  # 删除并返回值
del battle_data["damage_taken"]                # 直接删除

print(f"删除deaths后: {battle_data}")
print(f"被删除的deaths值: {removed_deaths}")

# 字典遍历 - 真实机器人数据
print(f"\n=== 字典遍历 ===")
robot_team = {
    "Hero": {"hp": 600, "damage": 10, "role": "重火力输出", "weapon": "42mm大弹丸"},
    "Infantry-1": {"hp": 200, "damage": 1, "role": "快速突击", "weapon": "17mm小弹丸"},
    "Infantry-2": {"hp": 200, "damage": 1, "role": "快速突击", "weapon": "17mm小弹丸"},
    "Engineer": {"hp": 250, "damage": 0, "role": "后勤支援", "weapon": "兑矿装置"}
}

print("RoboMaster队伍配置:")
for robot_name, config in robot_team.items():
    print(f"  {robot_name}: {config['hp']}HP, {config['damage']}攻击, {config['role']}, 装备: {config['weapon']}")

# 统计信息
total_hp = sum(config["hp"] for config in robot_team.values())
total_damage = sum(config["damage"] for config in robot_team.values())
combat_robots = len([r for r in robot_team.values() if r["damage"] > 0])

print(f"\n队伍总血量: {total_hp}")
print(f"队伍总攻击力: {total_damage}")
print(f"战斗机器人数量: {combat_robots}")

# 嵌套字典（复杂数据结构）
print(f"\n=== 嵌套字典 ===")
tournament_data = {
    "2024": {
        "teams": ["清华", "北理", "华科", "交大"],
        "champion": "清华",
        "robot_rules": {
            "Hero": {"max_hp": 600, "weapon_damage": 10},
            "Infantry": {"max_hp": 200, "weapon_damage": 1},
            "Engineer": {"max_hp": 250, "weapon_damage": 0}
        },
        "matches": {
            "semi_final": ["清华 vs 北理", "华科 vs 交大"],
            "final": "清华 vs 华科"
        }
    }
}

print(f"2024年参赛队伍: {tournament_data['2024']['teams']}")
print(f"冠军: {tournament_data['2024']['champion']}")
print(f"英雄机器人数据: {tournament_data['2024']['robot_rules']['Hero']}")
print(f"决赛: {tournament_data['2024']['matches']['final']}")

# 字典推导式
print(f"\n=== 字典推导式 ===")
robot_types = ["Hero", "Infantry", "Engineer"]
hp_values = [600, 200, 250]
damage_values = [10, 1, 0]

# 创建机器人规格字典
robot_specs = {robot: {"hp": hp, "damage": damage} 
               for robot, hp, damage in zip(robot_types, hp_values, damage_values)}

# 筛选可攻击的机器人
combat_robots_dict = {name: specs for name, specs in robot_specs.items() if specs["damage"] > 0}

print(f"机器人规格: {robot_specs}")
print(f"战斗机器人: {combat_robots_dict}")

# 实际应用：机器人状态管理（使用函数和字典）
print(f"\n=== RoboMaster 应用示例 ===")

# 使用字典存储机器人数据
robots_database = {}

def add_robot(robot_id, name, robot_type):
    """添加机器人到数据库"""
    # 根据类型设置规格
    if robot_type == "Hero":
        specs = {"hp": 600, "max_hp": 600, "damage": 10, "weapon": "42mm大弹丸"}
    elif robot_type == "Infantry":
        specs = {"hp": 200, "max_hp": 200, "damage": 1, "weapon": "17mm小弹丸"}
    elif robot_type == "Engineer":
        specs = {"hp": 250, "max_hp": 250, "damage": 0, "weapon": "兑矿装置"}
    else:
        print(f"❌ 未知机器人类型: {robot_type}")
        return False
    
    robots_database[robot_id] = {
        "name": name,
        "type": robot_type,
        **specs,  # 展开规格字典
        "status": "待命",
        "position": (0, 0)
    }
    print(f"✅ 已添加{robot_type}: {name} (ID: {robot_id})")
    return True

def attack_robot(attacker_id, target_id):
    """机器人攻击"""
    if attacker_id not in robots_database or target_id not in robots_database:
        print("❌ 机器人ID不存在")
        return False
    
    attacker = robots_database[attacker_id]
    target = robots_database[target_id]
    
    if attacker["damage"] == 0:
        print(f"❌ {attacker['name']} 无攻击能力")
        return False
    
    if target["hp"] <= 0:
        print(f"❌ {target['name']} 已被摧毁")
        return False
    
    # 执行攻击
    damage = attacker["damage"]
    target["hp"] = max(0, target["hp"] - damage)
    
    print(f"� {attacker['name']} 攻击 {target['name']}, 造成 {damage} 点伤害")
    print(f"   {target['name']} 血量: {target['hp']}/{target['max_hp']}")
    
    if target["hp"] == 0:
        target["status"] = "损坏"
        print(f"💀 {target['name']} 被摧毁!")
    
    return True

def get_robots_status():
    """显示所有机器人状态"""
    print("🤖 机器人状态总览:")
    for robot_id, info in robots_database.items():
        status_icon = "🟢" if info["status"] == "待命" else "🔴"
        print(f"  {status_icon} {robot_id}: {info['name']} ({info['type']}) - "
              f"{info['hp']}/{info['max_hp']}HP, {info['damage']}攻击, {info['weapon']}")

# 演示机器人管理系统
print("创建RoboMaster机器人队伍:")
add_robot("H001", "Hero-Alpha", "Hero")
add_robot("I001", "Infantry-Beta", "Infantry")
add_robot("I002", "Infantry-Gamma", "Infantry") 
add_robot("E001", "Engineer-Delta", "Engineer")

print("\n初始状态:")
get_robots_status()

# 模拟战斗
print(f"\n⚔️ 模拟战斗:")
attack_robot("H001", "I001")  # 英雄攻击步兵
attack_robot("I002", "H001")  # 步兵攻击英雄
attack_robot("E001", "I001")  # 工程试图攻击（失败）

print("\n战斗后状态:")
get_robots_status()

# 机器人数据统计
print(f"\n📊 队伍统计:")
total_robots = len(robots_database)
active_robots = len([r for r in robots_database.values() if r["status"] == "待命"])
total_hp = sum(r["hp"] for r in robots_database.values())
total_max_hp = sum(r["max_hp"] for r in robots_database.values())
hp_percentage = (total_hp / total_max_hp) * 100

print(f"  总机器人数: {total_robots}")
print(f"  战斗可用: {active_robots}")
print(f"  整体血量: {total_hp}/{total_max_hp} ({hp_percentage:.1f}%)")

# 找出最强机器人
if robots_database:
    strongest = max(robots_database.values(), key=lambda r: r["damage"])
    healthiest = max(robots_database.values(), key=lambda r: r["hp"])
    print(f"  攻击最强: {strongest['name']} ({strongest['damage']}点伤害)")
    print(f"  血量最高: {healthiest['name']} ({healthiest['hp']}HP)")

print("\n✅ 字典操作学习完成！")
```

### 2.1 条件语句 - 让程序做决策

条件语句让程序能够根据不同的情况执行不同的代码，这是编程逻辑的核心。

```python
# 条件语句详解

print("=== 基本条件语句 ===")

# 简单if语句
robot_hp = 150

if robot_hp > 200:
    status = "状态优秀"
    action = "继续战斗"
elif robot_hp > 100:
    status = "状态良好" 
    action = "小心应战"
elif robot_hp > 50:
    status = "状态一般"
    action = "寻找掩护"
else:
    status = "状态危急"
    action = "立即撤退"

print(f"机器人血量: {robot_hp}")
print(f"状态评估: {status}")
print(f"建议行动: {action}")

# 复杂条件判断
print(f"\\n=== 复杂条件判断 ===")

def analyze_robot_status(hp, armor, ammo, energy):
    """分析机器人综合状态"""
    
    print(f"分析参数: HP={hp}, 护甲={armor}, 弹药={ammo}, 能量={energy}")
    
    # 多条件组合判断
    if hp <= 0:
        return "机器人已损坏", "需要维修"
    
    if hp > 150 and armor > 50 and ammo > 100 and energy > 70:
        return "战斗状态极佳", "主动出击"
    elif hp > 100 and (armor > 30 or ammo > 50) and energy > 50:
        return "战斗状态良好", "谨慎推进"
    elif hp > 50 and energy > 30:
        if ammo <= 20:
            return "弹药不足", "补充弹药"
        elif armor <= 20:
            return "护甲受损", "维修护甲"
        else:
            return "状态一般", "保守作战"
    else:
        return "状态危急", "立即撤退"

# 测试不同状态
test_cases = [
    (200, 80, 150, 90),  # 最佳状态
    (120, 40, 60, 65),   # 良好状态
    (80, 15, 30, 45),    # 一般状态
    (30, 5, 10, 20),     # 危急状态
    (0, 0, 0, 0)         # 损坏状态
]

for i, (hp, armor, ammo, energy) in enumerate(test_cases, 1):
    status, action = analyze_robot_status(hp, armor, ammo, energy)
    print(f"测试{i}: {status} -> {action}")
    print()

# 三元运算符（条件表达式）
print(f"=== 三元运算符 ===")
robot_energy = 75
battery_status = "充足" if robot_energy > 50 else "不足"
print(f"能量: {robot_energy}%, 电池状态: {battery_status}")

# 链式比较
temperature = 25
comfort_level = (
    "过热" if temperature > 30 else
    "舒适" if 20 <= temperature <= 30 else
    "过冷"
)
print(f"温度: {temperature}°C, 舒适度: {comfort_level}")

# match语句（Python 3.10+的新特性）
print(f"\\n=== match语句示例 ===")

def handle_robot_command(command):
    """处理机器人指令"""
    
    # 注意：match语句需要Python 3.10+
    # 这里用if-elif模拟match的效果
    
    if command == "attack":
        return "执行攻击指令"
    elif command == "defend":
        return "进入防御模式"
    elif command == "retreat":
        return "开始撤退"
    elif command in ["move_forward", "move_backward", "move_left", "move_right"]:
        return f"执行移动指令: {command}"
    elif command.startswith("target_"):
        target = command.split("_")[1]
        return f"锁定目标: {target}"
    else:
        return "未知指令，请重新输入"

# 测试指令处理
commands = ["attack", "defend", "move_forward", "target_enemy1", "invalid_cmd"]
for cmd in commands:
    result = handle_robot_command(cmd)
    print(f"指令 '{cmd}': {result}")

# 条件语句的最佳实践
print(f"\\n=== 条件语句最佳实践 ===")

def evaluate_team_performance(scores) -> dict:
    """评估队伍表现"""
    
    if not scores:  # 检查空列表
        return {
            "performance": "",
            "average": 0.0,
            "range": 0,
            "consistency": "未知"
        }
    
    avg_score = sum(scores) / len(scores)
    max_score = max(scores)
    min_score = min(scores)
    
    # 使用常量提高代码可读性
    EXCELLENT_THRESHOLD = 90
    GOOD_THRESHOLD = 75
    AVERAGE_THRESHOLD = 60
    
    if avg_score >= EXCELLENT_THRESHOLD:
        performance = "优秀"
    elif avg_score >= GOOD_THRESHOLD:
        performance = "良好"
    elif avg_score >= AVERAGE_THRESHOLD:
        performance = "一般"
    else:
        performance = "需要改进"
    
    # 额外分析
    score_range = max_score - min_score
    consistency = "稳定" if score_range <= 20 else "不稳定"
    
    return {
        "performance": performance,
        "average": round(avg_score, 1),
        "range": score_range,
        "consistency": consistency
    }

# 测试队伍表现评估
team_scores = [88, 92, 85, 90, 87]
evaluation = evaluate_team_performance(team_scores)

print(f"队伍分数: {team_scores}")
print(f"表现评估: {evaluation['performance']}")
print(f"平均分: {evaluation['average']}")
print(f"分数波动: {evaluation['range']} (状态: {evaluation['consistency']})")

print("\\n✅ 条件语句学习完成！")
```

### 2.2 循环语句 - 重复执行任务

循环让程序能够重复执行相同的操作，这在处理大量数据或重复性任务时非常有用。

```python
# 循环语句详解

print("=== for循环 - 遍历序列 ===")

# 基本for循环
robots = ["Hero", "Infantry", "Engineer", "Drone"]
print("检查机器人状态:")
for robot in robots:
    print(f"  正在检查 {robot}...")

# 遍历索引和值
print(f"\\n带索引的遍历:")
for index, robot in enumerate(robots):
    print(f"  {index + 1}. {robot}")

# 遍历字典
robot_hp = {"Hero": 400, "Infantry": 200, "Engineer": 250, "Drone": 100}
print(f"\\n机器人血量统计:")

for robot_name, hp in robot_hp.items():
    print(f"  {robot_name}: {hp} HP")

# range()函数的使用
print(f"\\n=== range()函数 ===")

# 基本range
print("倒计时:")
for i in range(5, 0, -1):
    print(f"  {i}...")
print("  发射!")

# 生成数字序列
print(f"\\n偶数序列 (0-10):")
even_numbers = []
for i in range(0, 11, 2):
    even_numbers.append(i)
print(f"  {even_numbers}")

# 复杂for循环应用
print(f"\\n=== 复杂for循环应用 ===")

# 嵌套循环 - 生成战斗配对
teams = ["红队", "蓝队", "绿队"]
robots_per_team = ["Hero", "Infantry", "Engineer"]

print("比赛对阵表:")
match_count = 1
for i in range(len(teams)):
    for j in range(i + 1, len(teams)):
        print(f"  第{match_count}场: {teams[i]} vs {teams[j]}")
        match_count += 1

# 列表推导式（for循环的简化写法）
print(f"\\n列表推导式示例:")
numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
even_squares = [x**2 for x in numbers if x % 2 == 0]

print(f"原数字: {numbers}")
print(f"平方数: {squares}")
print(f"偶数的平方: {even_squares}")

# while循环
print(f"\\n=== while循环 - 条件循环 ===")

# 基本while循环
robot_energy = 100
round_num = 1

print("机器人能量消耗模拟:")
while robot_energy > 20:
    energy_cost = 15
    robot_energy -= energy_cost
    print(f"  第{round_num}轮: 消耗{energy_cost}能量, 剩余{robot_energy}能量")
    round_num += 1

print(f"能量不足，需要充电!")

# while循环的实际应用
print(f"\\n=== 用户输入验证 ===")

def simulate_input_validation():
    """模拟用户输入验证过程"""
    
    # 模拟用户输入序列
    simulated_inputs = ["abc", "-5", "150", "25"]
    input_index = 0
    
    print("请输入机器人血量 (1-100):")
    
    while input_index < len(simulated_inputs):
        user_input = simulated_inputs[input_index]
        print(f"输入: {user_input}")
        input_index += 1
        
        try:
            hp = int(user_input)
            if 1 <= hp <= 100:
                print(f"✅ 有效输入: {hp}")
                return hp
            else:
                print("❌ 血量必须在1-100之间，请重新输入:")
        except ValueError:
            print("❌ 请输入有效数字，重新输入:")
    
    print("输入次数过多，使用默认值50")
    return 50

validated_hp = simulate_input_validation()
print(f"最终血量设置: {validated_hp}")

# 循环控制语句
print(f"\\n=== 循环控制语句 ===")

# break语句 - 提前退出循环
scores = [95, 88, 72, 45, 91, 67]
print(f"分数列表: {scores}")
print("查找第一个不及格分数:")

for i, score in enumerate(scores):
    print(f"  检查第{i+1}个分数: {score}")
    if score < 60:
        print(f"  ❌ 找到不及格分数: {score}")
        break
else:
    print("  ✅ 所有分数都及格")

# continue语句 - 跳过当前迭代
print(f"\\n计算及格分数的平均值:")
passing_scores = []
for score in scores:
    if score < 60:
        print(f"  跳过不及格分数: {score}")
        continue
    passing_scores.append(score)
    print(f"  添加及格分数: {score}")

if passing_scores:
    avg = sum(passing_scores) / len(passing_scores)
    print(f"及格分数平均值: {avg:.1f}")

# 实际应用：机器人巡逻系统
print(f"\\n=== RoboMaster应用：机器人巡逻系统 ===")

def robot_patrol_simulation():
    """机器人巡逻模拟"""
    
    patrol_points = [
        (0, 0, "起点"),
        (10, 0, "检查点A"), 
        (10, 10, "检查点B"),
        (0, 10, "检查点C"),
        (0, 0, "终点")
    ]
    
    robot_energy = 100
    current_position = (0, 0)
    
    print("开始巡逻任务:")
    
    for i, (x, y, point_name) in enumerate(patrol_points):
        # 计算移动距离
        distance = ((x - current_position[0])**2 + (y - current_position[1])**2)**0.5
        energy_needed = distance * 2  # 每单位距离消耗2点能量
        
        print(f"\\n目标: {point_name} ({x}, {y})")
        print(f"距离: {distance:.1f}m, 需要能量: {energy_needed:.1f}")
        
        if robot_energy < energy_needed:
            print("❌ 能量不足，巡逻任务中断!")
            break
        
        robot_energy -= energy_needed
        current_position = (x, y)
        
        print(f"✅ 到达 {point_name}, 剩余能量: {robot_energy:.1f}")
        
        # 检查点特殊操作
        if "检查点" in point_name:
            print(f"   执行检查任务...")
            robot_energy -= 5  # 检查任务消耗5点能量
            print(f"   检查完成，剩余能量: {robot_energy:.1f}")
    
    else:
        print(f"\\n🎉 巡逻任务圆满完成!")
    
    return current_position, robot_energy

final_pos, final_energy = robot_patrol_simulation()
print(f"\\n最终位置: {final_pos}, 最终能量: {final_energy:.1f}")

print("\\n✅ 循环语句学习完成！")
```

## 第三部分：函数进阶（25 分钟）

### 3.1 函数定义和参数

函数是代码复用的重要工具，让我们能够将复杂的任务分解为简单的、可重复使用的模块。

```python
# 函数进阶应用

print("=== 函数定义和参数 ===")

# 基本函数定义
def calculate_damage(attack_power, armor_defense):
    """
    计算实际伤害值
    
    Args:
        attack_power (int): 攻击力
        armor_defense (int): 护甲防御力
    
    Returns:
        int: 实际造成的伤害
    """
    damage = max(1, attack_power - armor_defense)  # 至少造成1点伤害
    return damage

# 测试基本函数
damage1 = calculate_damage(50, 20)
damage2 = calculate_damage(30, 35)
print(f"攻击力50 vs 护甲20 = {damage1}点伤害")
print(f"攻击力30 vs 护甲35 = {damage2}点伤害")

# 默认参数
def create_robot(name, robot_type="Infantry", hp=200, speed=3.0):
    """
    创建机器人对象
    
    Args:
        name (str): 机器人名称
        robot_type (str): 机器人类型，默认为Infantry
        hp (int): 生命值，默认为200
        speed (float): 移动速度，默认为3.0
    
    Returns:
        dict: 机器人信息字典
    """
    robot = {
        "name": name,
        "type": robot_type,
        "hp": hp,
        "max_hp": hp,
        "speed": speed,
        "status": "待命"
    }
    return robot

# 测试默认参数
print(f"\\n=== 默认参数测试 ===")
robot1 = create_robot("普通步兵")
robot2 = create_robot("英雄机器人", "Hero", 400, 2.5)
robot3 = create_robot("工程机器人", robot_type="Engineer", hp=250)

print(f"机器人1: {robot1}")
print(f"机器人2: {robot2}")
print(f"机器人3: {robot3}")

# 可变参数 (*args)
def calculate_total_score(*scores):
    """
    计算总分和平均分
    
    Args:
        *scores: 可变数量的分数
    
    Returns:
        tuple: (总分, 平均分, 分数个数)
    """
    if not scores:
        return 0, 0, 0
    
    total = sum(scores)
    average = total / len(scores)
    return total, average, len(scores)

# 测试可变参数
print(f"\\n=== 可变参数测试 ===")
result1 = calculate_total_score(85, 92, 78, 90)
result2 = calculate_total_score(95, 88, 76, 89, 91, 84)

print(f"4个分数: 总分={result1[0]}, 平均={result1[1]:.1f}, 个数={result1[2]}")
print(f"6个分数: 总分={result2[0]}, 平均={result2[1]:.1f}, 个数={result2[2]}")

# 关键字参数 (**kwargs)
def create_team(**members):
    """
    创建队伍
    
    Args:
        **members: 队员信息的关键字参数
    
    Returns:
        dict: 队伍信息
    """
    team = {
        "members": members,
        "size": len(members),
        "captain": members.get("captain", "未指定")
    }
    return team

# 测试关键字参数
print(f"\\n=== 关键字参数测试 ===")
team1 = create_team(captain="张三", programmer="李四", operator="王五")
team2 = create_team(leader="赵六", member1="孙七", member2="周八", member3="吴九")

print(f"队伍1: {team1}")
print(f"队伍2: {team2}")

# 混合参数类型
def advanced_robot_config(name, robot_type, *weapons, hp=200, **special_abilities):
    """
    高级机器人配置函数
    
    Args:
        name (str): 机器人名称
        robot_type (str): 机器人类型
        *weapons: 武器列表
        hp (int): 生命值
        **special_abilities: 特殊能力
    
    Returns:
        dict: 完整的机器人配置
    """
    config = {
        "name": name,
        "type": robot_type,
        "hp": hp,
        "weapons": list(weapons),
        "abilities": special_abilities
    }
    return config

# 测试混合参数
print(f"\\n=== 混合参数测试 ===")
hero_config = advanced_robot_config(
    "超级英雄", "Hero",
    "17mm机枪", "42mm大炮",
    hp=400,
    shield=True,
    boost_speed=True,
    special_attack="能量炮"
)

print(f"英雄配置: {hero_config}")

# 函数作为参数
def apply_strategy(robots, strategy_func):
    """
    对机器人列表应用策略函数
    
    Args:
        robots (list): 机器人列表
        strategy_func (function): 策略函数
    
    Returns:
        list: 应用策略后的结果
    """
    return [strategy_func(robot) for robot in robots]

def attack_strategy(robot):
    """攻击策略"""
    return f"{robot['name']} 执行攻击任务"

def defense_strategy(robot):
    """防御策略"""
    return f"{robot['name']} 执行防御任务"

# 测试函数作为参数
robots = [
    {"name": "Hero-1", "type": "Hero"},
    {"name": "Infantry-1", "type": "Infantry"},
    {"name": "Engineer-1", "type": "Engineer"}
]

print(f"\\n=== 函数作为参数测试 ===")
attack_results = apply_strategy(robots, attack_strategy)
defense_results = apply_strategy(robots, defense_strategy)

print("攻击策略结果:")
for result in attack_results:
    print(f"  {result}")

print("防御策略结果:")
for result in defense_results:
    print(f"  {result}")

# 闭包和装饰器预览
print(f"\\n=== 高级特性预览 ===")

def create_robot_factory(robot_type):
    """
    创建机器人工厂函数（闭包示例）
    
    Args:
        robot_type (str): 机器人类型
    
    Returns:
        function: 创建特定类型机器人的函数
    """
    def create_specific_robot(name, hp=200):
        return {
            "name": name,
            "type": robot_type,
            "hp": hp
        }
    return create_specific_robot

# 创建专门的工厂函数
hero_factory = create_robot_factory("Hero")
infantry_factory = create_robot_factory("Infantry")

# 使用工厂函数
hero1 = hero_factory("Hero-Alpha", 400)
infantry1 = infantry_factory("Infantry-Beta", 200)

print(f"英雄机器人: {hero1}")
print(f"步兵机器人: {infantry1}")

# 递归函数示例
def calculate_fibonacci(n):
    """
    计算斐波那契数列
    
    Args:
        n (int): 序列位置
    
    Returns:
        int: 斐波那契数值
    """
    if n <= 1:
        return n
    return calculate_fibonacci(n-1) + calculate_fibonacci(n-2)

print(f"\\n=== 递归函数示例 ===")
fib_sequence = [calculate_fibonacci(i) for i in range(10)]
print(f"斐波那契数列前10项: {fib_sequence}")

print("\\n✅ 函数进阶学习完成！")
```

```python
# Lesson-2 综合实践项目：学生成绩管理系统

print("🎓 学生成绩管理系统")
print("=" * 50)

# 使用全局变量存储数据
students = {}  # 存储学生信息
courses = []   # 存储课程列表

def add_course(course_name):
    """添加课程"""
    global courses
    if course_name not in courses:
        courses.append(course_name)
        print(f"✅ 已添加课程: {course_name}")
    else:
        print(f"❌ 课程 {course_name} 已存在")

def add_student(student_id, name):
    """添加学生"""
    global students
    if student_id not in students:
        students[student_id] = {
            "name": name,
            "grades": {}
        }
        print(f"✅ 已添加学生: {name} (ID: {student_id})")
    else:
        print(f"❌ 学生ID {student_id} 已存在")

def add_grade(student_id, course, score):
    """添加成绩"""
    global students, courses
    if student_id in students:
        if course in courses:
            students[student_id]["grades"][course] = score
            print(f"✅ 已录入 {students[student_id]['name']} 的 {course} 成绩: {score}")
        else:
            print(f"❌ 课程 {course} 不存在，请先添加课程")
    else:
        print(f"❌ 学生ID {student_id} 不存在")

def get_student_average(student_id):
    """计算学生平均分"""
    global students
    if student_id in students:
        grades = list(students[student_id]["grades"].values())
        if grades:
            return sum(grades) / len(grades)
        else:
            return 0
    return None

def get_course_statistics(course):
    """获取课程统计信息"""
    global students
    scores = []
    for student in students.values():
        if course in student["grades"]:
            scores.append(student["grades"][course])
    
    if scores:
        return {
            "count": len(scores),
            "average": sum(scores) / len(scores),
            "max": max(scores),
            "min": min(scores),
            "pass_rate": len([s for s in scores if s >= 60]) / len(scores) * 100
        }
    return None

def display_all_students():
    """显示所有学生信息"""
    global students
    print(f"\n📊 学生信息总览:")
    print("-" * 60)
    
    for student_id, info in students.items():
        name = info["name"]
        grades = info["grades"]
        avg = get_student_average(student_id)
        
        print(f"学号: {student_id} | 姓名: {name} | 平均分: {avg:.1f}")
        
        if grades:
            for course, score in grades.items():
                status = "✅ 及格" if score >= 60 else "❌ 不及格"
                print(f"  └─ {course}: {score}分 {status}")
        else:
            print("  └─ 暂无成绩记录")
        print()

def display_course_statistics():
    """显示课程统计"""
    global courses
    print(f"\n📈 课程统计信息:")
    print("-" * 60)
    
    for course in courses:
        stats = get_course_statistics(course)
        if stats:
            print(f"课程: {course}")
            print(f"  参与人数: {stats['count']}")
            print(f"  平均分: {stats['average']:.1f}")
            print(f"  最高分: {stats['max']}")
            print(f"  最低分: {stats['min']}")
            print(f"  及格率: {stats['pass_rate']:.1f}%")
            print()
        else:
            print(f"课程: {course} (暂无成绩数据)\n")

def reset_system():
    """重置系统数据"""
    global students, courses
    students = {}
    courses = []
    print("🔄 系统数据已重置")

def demo_grade_system():
    """演示成绩管理系统"""
    
    # 重置系统
    reset_system()
    
    # 添加课程
    print("1️⃣ 添加课程")
    course_list = ["Python编程", "数据结构", "算法设计", "机器学习"]
    for course in course_list:
        add_course(course)
    
    print(f"\n2️⃣ 添加学生")
    # 添加学生
    student_list = [
        ("2024001", "张三"),
        ("2024002", "李四"), 
        ("2024003", "王五"),
        ("2024004", "赵六"),
        ("2024005", "孙七")
    ]
    
    for student_id, name in student_list:
        add_student(student_id, name)
    
    print(f"\n3️⃣ 录入成绩")
    # 录入成绩
    grades_data = [
        ("2024001", "Python编程", 95),
        ("2024001", "数据结构", 88),
        ("2024001", "算法设计", 92),
        ("2024002", "Python编程", 78),
        ("2024002", "数据结构", 85),
        ("2024002", "机器学习", 90),
        ("2024003", "Python编程", 56),
        ("2024003", "数据结构", 72),
        ("2024003", "算法设计", 68),
        ("2024004", "Python编程", 91),
        ("2024004", "机器学习", 87),
        ("2024005", "Python编程", 83),
        ("2024005", "数据结构", 79),
        ("2024005", "算法设计", 85),
        ("2024005", "机器学习", 92)
    ]
    
    for student_id, course, score in grades_data:
        add_grade(student_id, course, score)
    
    # 显示结果
    display_all_students()
    display_course_statistics()

# 运行演示
demo_grade_system()

print("\n" + "=" * 50)
print("🎉 学生成绩管理系统演示完成!")
print("=" * 50)

# 额外功能演示
print(f"\n📚 系统功能演示:")

# 查询单个学生成绩
print("\n� 查询功能演示:")
student_to_check = "2024001"
if student_to_check in students:
    student_info = students[student_to_check]
    avg_score = get_student_average(student_to_check)
    print(f"学生 {student_info['name']} (ID: {student_to_check}):")
    print(f"  平均分: {avg_score:.1f}")
    for course, score in student_info['grades'].items():
        print(f"  {course}: {score}分")

# 查询课程信息
print(f"\n📊 课程分析:")
course_to_analyze = "Python编程"
stats = get_course_statistics(course_to_analyze)
if stats:
    print(f"{course_to_analyze} 课程分析:")
    print(f"  参与学生: {stats['count']} 人")
    print(f"  平均分: {stats['average']:.1f}")
    print(f"  最高分: {stats['max']}")
    print(f"  最低分: {stats['min']}")
    print(f"  及格率: {stats['pass_rate']:.1f}%")

# 数据统计
print(f"\n📈 整体统计:")
total_students = len(students)
total_courses = len(courses)
total_grades = sum(len(student['grades']) for student in students.values())

print(f"  学生总数: {total_students}")
print(f"  课程总数: {total_courses}")
print(f"  成绩记录总数: {total_grades}")

# 找出最优秀的学生
if students:
    best_student_id = None
    best_average = 0
    
    for student_id in students:
        avg = get_student_average(student_id)
        if avg and avg > best_average:
            best_average = avg
            best_student_id = student_id
    
    if best_student_id:
        best_student_name = students[best_student_id]['name']
        print(f"  最优秀学生: {best_student_name} (平均分: {best_average:.1f})")

print(f"\n📚 本项目使用的知识点:")
print("✅ 字典存储复杂数据结构")
print("✅ 列表管理集合数据")
print("✅ 函数封装功能模块")
print("✅ 全局变量管理程序状态")
print("✅ for循环遍历数据")
print("✅ 条件语句进行逻辑判断")
print("✅ 字符串格式化输出")
print("✅ 数学统计计算")
print("✅ 字典方法：keys(), values(), items()")
print("✅ 列表推导式进行数据过滤")

print(f"\n🚀 恭喜完成 Lesson-2 的学习!")
print("下节课我们将学习面向对象编程，让程序结构更加清晰和强大！")
```
